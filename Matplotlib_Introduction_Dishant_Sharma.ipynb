```python
hobbies=["Sports","Coding","Cooking","Dancing","Gardening"]
grp_1=[30,10,20,25,10]
grp_2=[40,5,20,20,10]
grp_3=[35,5,30,15,15]
```

```python
plt.barh(hobbies,grp_2,color="c") #horizontal bar
```

```python
plt.bar(hobbies,grp_3,width=0.3,align="edge",color="g",edgecolor="y",linewidth=5,linestyle="--")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(hobbies,grp_1,width=0.2,color="b")
plt.bar(hobbies,grp_2,width=0.2,color="g")

plt.title("General Hobbies")
plt.xlabel("Hobby")
plt.ylabel("No. of people")
```

```python
plt.figure(figsize=(16,9))
width=0.2
hobbies_index=np.arange(len(hobbies))
plt.bar(hobbies_index,grp_1,width,color="b")
plt.bar(hobbies_index+width,grp_2,width,color="g")
plt.bar(hobbies_index+width+width,grp_3,width,color="y")

plt.xticks(hobbies_index+width,hobbies)

plt.title("General Hobbies")
plt.xlabel("Hobby")
plt.ylabel("No. of people")
```

4)Box Plot using Matplotlib

Boxplot is a way of representing the distribution of the data based on the "FIVE POINT SUMMARY"

The five number summary includes 5 items:

1) The minimum.
2) Q1 (the first quartile, or the 25% mark).
3) The median.
4) Q3 (the third quartile, or the 75% mark).
5) The maximum.

```python
Data=np.array([0,21,31,51,61,71,110,114,116,191,213])
```

```python
np.median(Data)
```

```python
quartile = np.percentile(Data, [25, 50, 75])
min, max = Data.min(), Data.max()

# print 5-number summary
print('Min: %.3f' % min)
print('Q1: %.3f' % quartile[0])
print('Median: %.3f' % quartile[1])
print('Q3: %.3f' % quartile[2])
print('Max: %.3f' % max)
```

```python
plt.boxplot(Data,vert=True,patch_artist=False);
```

Insights given by plotting a Boxplot
-> Visualizing the outliers
-> understand the curve/skewness
-> helps in identifying if the data is symmetrical

5) Pie Chart

```python
Maybe you want one of the wedges to stand out? 
The explode parameter allows you to do that.

The explode parameter, if specified, and not None,
must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:
```

```python
# Data to plot
labels = 'ML', 'DSA', 'WEB DEV', 'CLOUD COMPUTING'
sizes = [245, 130, 215, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.3, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%.2f%%', shadow=False)

plt.axis('equal')
plt.show()
```

Remark: We use autopct to display the percent value using Python string formatting.
For example, autopct='%.1f%%' means that for each pie wedge, the format string is '.1f%'.

6) Histograms

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

```python
import random
```

```python
cse_students_age=np.random.randint(18,45,(100))
ece_students_age=np.random.randint(15,40,(100))
```

```python
plt.hist(cse_students_age)

plt.title("CSE STUDENTS AGE")
plt.xlabel("AGE")
plt.ylabel("No.of students")
```

```python
bins=[15,20,25,30,35,40,45]
```

bins:How much range shoould a single bar cover  Bins define how many buckets to divide the data up into between intervals

```python
plt.hist(cse_students_age,bins)

plt.title("CSE STUDENTS AGE")
plt.xlabel("AGE")
plt.ylabel("No.of students")
```

rwidth,orientation

```python
plt.hist(cse_students_age,bins,rwidth=0.9,orientation='horizontal',color='m')

plt.title("CSE STUDENTS AGE")
plt.xlabel("AGE")
plt.ylabel("No.of students")
```

```python
plt.hist(cse_students_age,bins,rwidth=0.9,orientation='vertical',color='m',label="cse")
plt.hist(ece_students_age,bins,rwidth=0.9,orientation='vertical',color='g',label="ece")
plt.title("STUDENTS AGE")
plt.xlabel("AGE")
plt.ylabel("No.of students")
plt.legend()
```

```python
#pass differentiating parameters in a list
plt.hist([cse_students_age,ece_students_age],bins,rwidth=0.9,orientation='vertical',color=['m','b'],label=["cse","ece"])
plt.title("STUDENTS AGE")
plt.xlabel("AGE")
plt.ylabel("No.of students")
plt.legend()
```

6) Scatter plot

The scatter() function plots one dot for each observation.
It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
no_of_years_old_bike = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(no_of_years_old_bike,speed, color = 'hotpink')

no_of_years_old_bike = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
speed= np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(no_of_years_old_bike,speed, color = 'g')

plt.show()
```

By comparing the two plots,it is safe to say that they both gives us the same conclusion: the newer the car, the faster it drives.
